```python
import pickle as pkl
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import seaborn.snsplot
```

```python
with open('test_8_500_1000.pkl','rb') as f:
    df8 = pkl.load(f)
    
with open('test9_1000_1000.pkl','rb') as f:
    df9 = pkl.load(f)
with open('test9b_1000_2000.pkl','rb') as f:
    df9b = pkl.load(f)
```

```python
fig, axes = plt.subplots(1,2, figsize=(18, 5))
fig.suptitle('Results: MSE (ATE)')


sns.lineplot(ax=axes[0], x=test_sizes, y=df9['MSE1'],  marker='X', dashes = True, color = 'Orange')
sns.lineplot(ax=axes[0], x=test_sizes, y=df9['MSE2'],  marker='X', dashes = True, color = 'Blue')

sns.lineplot(ax=axes[1], x=test_sizes, y=df9b['MSE1'],  marker='X', dashes = True, color = 'Orange')
sns.lineplot(ax=axes[1], x=test_sizes, y=df9b['MSE2'],  marker='X', dashes = True, color = 'Blue')
axes[0].title.set_text('n = 1000, MC reps = 1000')
axes[1].title.set_text('n = 2000, MC reps = 1000')


axes[0].legend(['X1 > 0', 'X1 <0'], loc='lower right')
axes[1].legend(['X1 > 0', 'X1 <0'], loc='lower right')


axes[0].set(xlabel='Estimation Sample Proportion', ylabel='MSE (ATE)')
axes[1].set(xlabel='Estimation Sample Proportion', ylabel='MSE (ATE)')

fig.savefig("mseate.png")
```

```python
fig, axes = plt.subplots(1,2, figsize=(18, 5))
fig.suptitle('Results: BIAS (ATE)')


sns.lineplot(ax=axes[0], x=test_sizes, y=df9['BIAS1'],  marker='X', dashes = True, color = 'Orange')
sns.lineplot(ax=axes[0], x=test_sizes, y=df9['BIAS2'],  marker='X', dashes = True, color = 'Blue')

sns.lineplot(ax=axes[1], x=test_sizes, y=df9b['BIAS1'],  marker='X', dashes = True, color = 'Orange')
sns.lineplot(ax=axes[1], x=test_sizes, y=df9b['BIAS2'],  marker='X', dashes = True, color = 'Blue')
axes[0].title.set_text('n = 1000, MC reps = 1000')
axes[1].title.set_text('n = 2000, MC reps = 1000')



axes[0].legend(['X1 > 0', 'X1 <0'], loc='lower right')
axes[1].legend(['X1 > 0', 'X1 <0'], loc='lower right')


axes[0].set(xlabel='Estimation Sample Proportion', ylabel='BIAS (ATE)')
axes[1].set(xlabel='Estimation Sample Proportion', ylabel='BIAS (ATE)')

fig.savefig("bias.png")
```
